```python
import pandas as pd
```

```python
demographic_gdp_data = pd.read_csv("Demographic_GDP.csv")
demographic_disaster_data = pd.read_csv("Demographic_Disaster.csv")
demographic_government_data = pd.read_csv("Demographic_Government.csv")
```

```python
combined_data = demographic_disaster_data.merge(demographic_gdp_data, on=['Country name', 'Year'])
combined_data = combined_data.merge(demographic_government_data, on=['Country name', 'Year'])
```

```python
combined_data.head()
```

```python
# Check for any missing values
print(combined_data.isnull().sum())

combined_data = combined_data.dropna()
```
